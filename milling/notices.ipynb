# Notices on the Milling Project 

1/ Regarding the simulations 

   The physics of the milling is dictated by what happens in the rotor-stator gap, most obviously. Without a proper mesh convergence analysis, results obtained for the flow are not meaningful. Indeed, any change in the model coefficients (surface roughness, rotor speed, turbulence model coefficient) cannot be properly accessed without a proper (volumetric and boundary layer) meshing strategy. 

    Q1: How have ideas on how to generate the mesh evolved since last time we spoke? How do you decide spatial resolution to be sufficient? 
    Q2: How much manual work does mesh generation require? What happens in case the airgap is reduced? To what extend can pre-processing be automated? 
    Q3: Could you complement your wonderful version of previous slides with information on the mesh (number of cells, mesh quality) and the solver used (e.g. residual convergence of the simulations, CPU time)?
